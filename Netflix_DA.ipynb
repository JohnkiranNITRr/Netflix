# Exploratory Data Analysis (EDA)

```python
plt.figure(figsize = (8,5))
ax = sns.countplot(data = df, x = 'type')
plt.title('Distribution of Content Types')
plt.xlabel('Type')
plt.ylabel('Count')
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
plt.figure(figsize = (12, 6))
ax = sns.countplot(data = df, x= 'rating', order= df['rating'].value_counts().index)
plt.title('Distribution of Content Ratings')
plt.xlabel('Rating')
plt.ylabel('Count')
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
plt.figure(figsize = (12, 6))
ax = sns.histplot(data = df, x = 'release_year', bins = 30, kde = True)
plt.title('Distribution of Content by release Year')
plt.xlabel('Release_year')
plt.ylabel('Count')
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
# Plot number of movies produced by each country
plt.figure(figsize=(14, 8))
top_countries = df['country'].value_counts().nlargest(10).index
sns.countplot(data=df, y='country', order=top_countries)
plt.title('Number of Titles Produced by Each Country')
plt.xlabel('Count')
plt.ylabel('Country')
plt.show()
```

```python

top_directors = df['director'].str.split(', ', expand=True).stack().value_counts().nlargest(10)
top_actors = df['cast'].str.split(', ', expand=True).stack().value_counts().nlargest(10)

# Plot for directors
plt.figure(figsize=(12, 6))
sns.barplot(x=top_directors.values, y=top_directors.index)
plt.title('Top 10 Directors by Number of Titles')
plt.xlabel('No. of Titles')
plt.ylabel('Director')
plt.show()
```

```python
# Plot for actors
plt.figure(figsize=(12, 6))
sns.barplot(x=top_actors.values, y=top_actors.index)
plt.title('Top 10 Actors by Number of Titles')
plt.xlabel('Count')
plt.ylabel('Actor')
plt.show()
```

# Based on EDA insights are :

# 1. The number of Movies and TV shows has increased more rapidly in recent years (2010-2020)
# 2. The highest count is observed in the 2020s with 1423, indicating a trend of more content being released in recent years.
# 3. US and India followed by UK produce more movies and tv shows 
# 4. TV-MA having the highest rating count at 1189, followed by TV-14 with 917.
